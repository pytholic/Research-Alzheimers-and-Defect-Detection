```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('D:/LAB Project Data/1D signal/boundary/With OTSU adaptive canny threshold/mean_filter_new.png',0)
img = cv2.medianBlur(img,5)

ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,11,2)
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)

titles = ['Original Image', 'Global Thresholding (v = 127)',
            'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img, th1, th2, th3]

fig = plt.figure(figsize=(20,20))
for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
# Python program to illustrate 
# Otsu thresholding type on an image 
  
# organizing imports 
import cv2          
import numpy as np     
from matplotlib import pyplot as plt

# path to input image is specified and 
# image is loaded with imread command 
image1 = cv2.imread('D:/LAB Project Data/1D signal/boundary/With OTSU adaptive canny threshold/mean_filter_new.png',0) 
  
# cv2.cvtColor is applied over the 
# image input with applied parameters 
# to convert the image in grayscale 
#img = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY) 
  
# applying Otsu thresholding 
# as an extra flag in binary  
# thresholding      
ret, thresh1 = cv2.threshold(image1, 120, 255, cv2.THRESH_BINARY + 
                                            cv2.THRESH_OTSU)      
  
# the window showing output image          
# with the corresponding thresholding          
# techniques applied to the input image     
fig = plt.figure(figsize=(20,20))
plt.title('Otsu Threshold')
plt.imshow (thresh1, cmap='gray') 
plt.axis('on')
#cv2.imshow('Otsu Threshold', thresh1)          
       
'''# De-allocate any associated memory usage          
cv2.waitKey(0) 
cv2.destroyAllWindows() '''
```
