```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('composite patch/new_try_tuning/mean_filter.png',0)
img = cv2.flip (img, 0)
#edges = cv2.Canny(img,200,300) #for jehung data
edges = cv2.Canny(img,165,255) #100,180
fig = plt.figure(figsize=(20,20))
#plt.subplot(121),plt.imshow(img,cmap = 'gray')
#plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.imshow(edges, origin='lower', cmap = 'gray')
#plt.title('Edge Image'), 
#plt.xticks([]), plt.yticks([])

plt.show()
fig.savefig('composite patch/new_try_tuning/canny_boundary.png', bbox_inches='tight')
plt.close(fig)
```

```python
indices = np.where(edges != [0])
coordinates = zip(indices[0], indices[1])
```

```python
defected = list(coordinates)
```

```python
print(defected)
```

```python
import pandas as pd

df = pd.DataFrame(defected)
df.to_csv('jehung_data/new_try/defected_boundary_new.csv', index=False, header=False)
```
