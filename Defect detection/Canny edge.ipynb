```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('mean_filter.png',0)
edges = cv2.Canny(img,80,200)

fig = plt.figure(figsize=(20,20))
#plt.subplot(121),plt.imshow(img,cmap = 'gray')
#plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.plot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
#fig.savefig(('canny with mean filter.png'))
plt.close(fig)
```

```python
indices = np.where(edges != [0])
coordinates = zip(indices[0], indices[1])
```

```python
defected = list(coordinates)
```

```python
print(defected)
```

```python
import pandas as pd

df = pd.DataFrame(defected)
df.to_csv('defected_boundary.csv', index=False, header=False)
```
