```python
import pydicom as dicom
import matplotlib.pyplot as plt
from PIL import Image
import glob
import os
import imageio
import numpy as np
```

```python
CATEGORIES = ['AD', 'CN', 'MCI']
inDir = 'F:/LAB/Dimentia/data new/Dicom test data/'
outDir = 'F:/LAB/Dimentia/data new/test_middle_slices/'
```

```python
for categ in CATEGORIES:
    folders = os.listdir(inDir + categ + '/')
    print(len(folders))
```

```python
categ = CATEGORIES[2]
i = 1
#os.mkdir(outDir + categ + '/' + f'{i}')
folders = os.listdir(inDir + categ + '/')
for folder in folders:
    PATH = os.path.join(inDir, categ, folder)
    os.mkdir(outDir + categ + '/' + f'{i}')
    out_path = outDir + categ + '/' + f'{i}'
    images = [os.path.basename(x) for x in glob.glob(PATH + '/*.dcm')]
    for f in images:  
        ds = dicom.read_file(PATH + '/' + f) # read dicom image
        img = ds.pixel_array # get image array
        img = np.flip(img, axis=0)
        imageio.imwrite(out_path + '/' + f.replace('.dcm','.png'), img)
    i += 1
```
