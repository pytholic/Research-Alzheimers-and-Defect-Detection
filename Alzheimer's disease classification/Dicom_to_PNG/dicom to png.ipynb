```python
import pydicom as dicom
import matplotlib.pyplot as plt
from PIL import Image
import glob
import os
import imageio
import numpy as np
```

```python
inDir = 'F:/LAB/Dimentia/data new/Dicom train data/MCI dicom data/'
outDir = 'F:/LAB/Dimentia/data new/48 slices data/train/MCI/'
```

```python
folders = os.listdir(inDir)
for folder in folders:
    PATH = os.path.join(inDir, folder)
    images = [os.path.basename(x) for x in glob.glob(PATH + '/*.dcm')]
    for f in images:  
        ds = dicom.read_file(PATH + '/' + f) # read dicom image
        img = ds.pixel_array # get image array
        img = np.flip(img, axis=0)
        imageio.imwrite(outDir + f.replace('.dcm','.png'), img)
```
